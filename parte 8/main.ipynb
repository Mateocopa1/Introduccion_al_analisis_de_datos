```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datos.csv')
```

```python
anemicos = df['anaemia'][df['anaemia']==1].count()
no_anemicos = df['anaemia'][df['anaemia']==0].count()

diabeticos = df['diabetes'][df['diabetes']==1].count()
no_diabeticos = df['diabetes'][df['diabetes']==0].count()

fumadores = df['smoking'][df['smoking']==1].count()
no_fumadores = df['smoking'][df['smoking']==0].count()

muertos = df['DEATH_EVENT'][df['DEATH_EVENT']==1].count()
vivos = df['DEATH_EVENT'][df['DEATH_EVENT']==0].count()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 4, figsize=(20, 15))

categorias = ['Si','No']
valores = [anemicos , no_anemicos]
startangle = 90
axs[0].pie(valores, labels=categorias, startangle=startangle,autopct='%1.1f%%')
axs[0].set_title("Anemicos")

categorias = ['Si','No']
valores = [diabeticos , no_diabeticos]
startangle = 90
axs[1].pie(valores, labels=categorias, startangle=startangle,autopct='%1.1f%%')
axs[1].set_title("Diabeticos")

categorias = ['Si','No']
valores = [fumadores, no_fumadores]
startangle = 90
axs[2].pie(valores, labels=categorias, startangle=startangle,autopct='%1.1f%%')
axs[2].set_title("Fumadores")

categorias = ['Si','No']
valores = [muertos , vivos]
startangle = 90
axs[3].pie(valores, labels=categorias, startangle=startangle,autopct='%1.1f%%')
axs[3].set_title("Muertos")

plt.show()

```
